```python
import pickle
for fname in os.listdir(stock_values_folder):
    if fname[-4:]=='.csv':
        print (fname)
        file_path=os.path.join(stock_values_folder,fname)
        f=open(file_path,'r')
        lines = f.readlines()
        f.close()
        dic={}
        for line in lines:
            line=line.replace('\n','').split(',')
            dic[line[0]]=float(line[1])

        with open(pickle_folder + fname+'.pkl','wb') as pick:
            pickle.dump(dic, pick, protocol=pickle.HIGHEST_PROTOCOL)

```
